```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000, n_redundant=0, n_features=2, n_clusters_per_class=1, weights=[0.90], random_state=1)
```

```python
df1 = pd.DataFrame(x, columns=['feature1', 'feature2'])
df2 = pd.DataFrame(y, columns=['target'])
df = pd.concat([df1, df2], axis=1)
df.head()
```

```python
df['target'].value_counts()
```

```python
plt.scatter(df['feature1'], df['feature2'], c=df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
## transform the dataset using SMOTE
smote = SMOTE()
X_smote, y_smote = smote.fit_resample(df[['feature1', 'feature2']], df['target'])
```

```python
X_smote.shape, y_smote.shape
```

```python
df1 = pd.DataFrame(X_smote, columns=['feature1', 'feature2'])
df2 = pd.DataFrame(y_smote, columns=['target'])
df_smote = pd.concat([df1, df2], axis=1)
```

```python
plt.scatter(df_smote['feature1'], df_smote['feature2'], c=df_smote['target'])
```
