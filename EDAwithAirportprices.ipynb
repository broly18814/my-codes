```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# loading the airplane prices
df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.head()
```

```python
df.tail()
```

```python
# add the columns day month year with the help of date_of_journey column
df['day'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y').dt.day # explaination: pd.to_datetime converts the Date_of_Journey column to datetime format, then dt.day extracts the day from the date
df['month'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y').dt.month
df['year'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y').dt.year
df.head()
```

```python
# drop date_of_journey column
df.drop('Date_of_Journey', axis=1, inplace=True)
df.head()
```

```python
# drop Route column as it is not useful
df.drop('Route', axis=1, inplace=True)
df.head()
```

```python
# remove strings in Dep_Time and Arrival_Time columns and Total_Stops column
df['Dep_Time'] = df['Dep_Time'].str.split(' ').str[0] # explanation: str.split(' ') splits the string at space and str[0] takes the first part
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
df['Total_Stops'] = df['Total_Stops'].str.split(' ').str[0]
df.head()
```

```python
# replace the values in Total_Stops column with numerical values
df['Total_Stops'] = df['Total_Stops'].replace({'non-stop': 0})
df.head()
```

```python
# covert Duration into minutes
def convert_to_minutes(duration):
    if 'h' in duration and 'm' in duration:
        hours, minutes = duration.split('h')
        minutes = minutes.replace('m', '')
        total_minutes = int(hours) * 60 + int(minutes)
    elif 'h' in duration:
        hours = duration.replace('h', '')
        total_minutes = int(hours) * 60
    elif 'm' in duration:
        minutes = duration.replace('m', '')
        total_minutes = int(minutes)
    else:
        total_minutes = 0
    return total_minutes
df['Duration'] = df['Duration'].apply(convert_to_minutes)
df.head()
```

```python
# divide Dep_Time and Arrival_Time into hours and minutes
df['Dep_Hour'] = pd.to_datetime(df['Dep_Time'], format='%H:%M').dt.hour
df['Dep_Minute'] = pd.to_datetime(df['Dep_Time'], format='%H:%M').dt.minute
df['Arrival_Hour'] = pd.to_datetime(df['Arrival_Time'], format='%H:%M').dt.hour
df['Arrival_Minute'] = pd.to_datetime(df['Arrival_Time'], format='%H:%M').dt.minute
df.head()
```

```python
# drop dep_time and arrival_time columns
df.drop(['Dep_Time', 'Arrival_Time'], axis=1, inplace=True)
df.head()
```

```python
# lets arrange the columns in a better way
df = df[['Airline', 'Source', 'Destination', 'Total_Stops', 'Duration', 'day', 'month', 'year', 'Dep_Hour', 'Dep_Minute', 'Arrival_Hour', 'Arrival_Minute', 'Price']]
df.head()
```
