```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# read the algerian forest fires dataset
data = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
data.head()
```

```python
data.info()
```

```python
## Data Cleaning
data.isnull().sum()
```

```python
data[data.isnull().any(axis=1)]
```

```python
data.loc[:122, 'Region']=0
data.loc[122:, 'Region']=1
df=data
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
# lets reset the index
df.iloc[[122]]
```

```python
df = df.drop(index=122).reset_index(drop=True)
df.head()
```

```python
df.drop(index=0)
```

```python
df.columns
```

```python
# Fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
# change the required columns into as integer data type
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
df.info()
```

```python
# changing other columns to float data type
#[features for features in df.columns if df[features].dtype == 'O']
# explanation for above line: list comprehension to get all features which are of object data type
for features in df.columns:
    if df[features].dtype == 'O' and features != 'Classes':
        df[features] = df[features].astype(float)
df.info()
```

```python
df.head()
```

```python
## lest save this dataset
df.to_csv('Algerian_forest_fires_cleaned.csv', index=False)
```

```python
# Exploratory Data Analysis
df1 = df.drop(columns=['day', 'month', 'year'], axis=1)
df1.head()
```

```python
# encoding categorical features
df1['Classes'] = np.where(df1['Classes'].str.contains('not fire'), 0, 1)
df1.head()
```

```python
# density plot for all features
print(plt.style.available)
plt.style.use('classic')
df1.hist(bins=50, figsize=(15,10))
plt.show()
```

```python
## Percentage of Pie chart
fire_counts = df1['Classes'].value_counts()
labels = ['Not Fire', 'Fire']
plt.figure(figsize=(7,7))
plt.pie(fire_counts, labels=labels, autopct='%1.1f%%')
plt.title('Percentage of Fire vs Not Fire Instances')
plt.show()
```

```python
## Correlation
df1.corr()
```

```python
sns.heatmap(df1.corr(), annot=True, cmap='coolwarm', fmt='.2f')
```

```python
# Monthly fire analysis graph comparing number of fires and not fires with Region with seaborn for bejaia only
df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 0, 1)
plt.figure(figsize=(10,5))
sns.set_style('whitegrid')
sns.countplot(data=df[df['Region']==0], x='month', hue='Classes')
plt.title('Monthly Fire Analysis in Bejaia Region')
plt.xlabel('Month')
plt.ylabel('Number of Instances')
plt.legend(title='Classes', labels=['Not Fire', 'Fire'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.set_style('whitegrid')
sns.countplot(data=df[df['Region']==1], x='month', hue='Classes')
plt.title('Monthly Fire Analysis in Sidi Region')
plt.xlabel('Month')
plt.ylabel('Number of Instances')
plt.legend(title='Classes', labels=['Not Fire', 'Fire'])
plt.show()
```
