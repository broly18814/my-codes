```python
### Data Visualization with Matplotlib
%pip install matplotlib
### Reading Data from Different Sources
%pip install pandas numpy openpyxl requests
import pandas as pd
import numpy as np
from io import StringIO
import requests
import matplotlib.pyplot as plt
%matplotlib inline
%pip install seaborn
import seaborn as sns
```

```python
# plotting a simple line graph
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
# customizing plots with titles and labels and legends and colors and multiple lines and line styles and markers and line widths and figure sizes and subplots and saving plots to files
plt.figure(figsize=(10, 6))
plt.plot(x, np.sin(x), label='Sine Wave', color='blue', linestyle='-', marker='o', linewidth=2)
plt.plot(x, np.cos(x), label='Cosine Wave', color='red', linestyle='--', marker='x', linewidth=2)
plt.title('Sine and Cosine Waves')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.savefig('sine_cosine_waves.png')
plt.show()
```

```python
## Data Visualization with seaborn
# plotting a simple scatter plot
tips = sns.load_dataset('tips')
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', sizes=(20, 200))
plt.title('Scatter Plot of Tips vs Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```
