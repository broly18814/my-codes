```python
## reading data from url of wine.csv
url_wine = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
df_wine = pd.read_csv(url_wine ,sep=';')
df_wine.head()
```

```python
## reading data from html table of a bank website
url_bank = 'https://www.federalreserve.gov/releases/h15/'
# To fix the 403 error, we can use requests to get the page with a user-agent header
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}
response = requests.get(url_bank, headers=headers)
tables = pd.read_html(response.text)
len(tables)
tables[0]
```

```python
## reading data from online excel file
# The original URL is blocking requests. We will use a different sample excel file.
url_excel = 'https://go.microsoft.com/fwlink/?LinkID=521962'

# For .xlsx files, we need to specify the engine as 'openpyxl'.
%pip install openpyxl
# We can pass the URL directly to read_excel
df_excel = pd.read_excel(url_excel, engine='openpyxl')
df_excel.head()
```

```python
## saving the data with pickle format
df_excel.to_pickle('excel_data.pkl')
df_pickle = pd.read_pickle('excel_data.pkl')
df_pickle.head()
```
